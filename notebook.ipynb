```python
# Install DAFT package (run this in terminal if not already installed)
# pip install 'daft-pgm'

import daft
import matplotlib.pyplot as plt
import numpy as np

# Set up plotting parameters for better quality
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 300

print("DAFT package imported successfully!")
print("Ready to create beautiful DAGs!")
```

```python
import daft
import matplotlib.pyplot as plt

# Create the DAG for 1747 Understanding
pgm1 = daft.PGM(dpi=150)

# Nodes
pgm1.add_node("lemons", "Lemons\n(Intake)", 0, 0, aspect=2.5, scale=1.2,
              plot_params={'facecolor': 'lightyellow', 'edgecolor': 'gold', 'linewidth': 2, 'alpha': 0.9})
pgm1.add_node("scurvy", "Scurvy\nPrevention", 2, 0, aspect=2.5, scale=1.2,
              plot_params={'facecolor': 'lightcoral', 'edgecolor': 'darkred', 'linewidth': 2, 'alpha': 0.9})

# Edge
pgm1.add_edge("lemons", "scurvy")

# Render
pgm1.render()
plt.title("DAG 1: Simplified Lind's Model (1747)", fontsize=12)
plt.show()
```

```python
import daft
import matplotlib.pyplot as plt

# Create the DAG for Misguided Belief
pgm2 = daft.PGM(dpi=150)

# Nodes
pgm2.add_node("acid", "Acid", 0, 0, aspect=2.5, scale=1.2,
              plot_params={'facecolor': 'lightblue', 'edgecolor': 'navy', 'linewidth': 2, 'alpha': 0.9})
pgm2.add_node("bacteria", "Bacteria in food", 2, 0, aspect=2.5, scale=1.2,
              plot_params={'facecolor': 'lightgreen', 'edgecolor': 'green', 'linewidth': 2, 'alpha': 0.9})
pgm2.add_node("scurvy", "Scurvy", 4, 0, aspect=2.5, scale=1.2,
              plot_params={'facecolor': 'lightcoral', 'edgecolor': 'darkred', 'linewidth': 2, 'alpha': 0.9})

# Edges
pgm2.add_edge("acid", "bacteria")
pgm2.add_edge("bacteria", "scurvy")

# Render
pgm2.render()
plt.title("DAG 2: Updated Bacterial Theory (Late 19th Century)", fontsize=12)
plt.show()
```
